**Question 1.** How many stories were created during the period??

```python
items = []
id = 7917
```

```python
while True:
    response = requests.get(url + 'item/' + str(id) + '.json')
    item_json = response.json()
    
    item_time = int(item_json['time'])
    item_timestamp = datetime.fromtimestamp(item_time)
    item_type = item_json['type']
    
    print(f'time: {item_timestamp}, type: {item_type}')
    
    id += 1
          
    if item_time < time_from:
        continue
        
    if item_time > time_to:
        break
    
    if item_json['type'] == 'story':
        items.append(item_json)
        print(item_json)
```

**Question 2.** What was the length of the longest story?

```python
comments = {}

progress_count = 0

for item in items:
    
    progress_count += 1
    progress = progress_count / len(items)
    print(f'Progress: {progress}')
    
    if 'kids' in item:
        
        comments_stack = item['kids']
        
        while True:
            if len(comments_stack) > 0:
                response = requests.get(url + 'item/' + str(comments_stack.pop()) + '.json')
                item_json = response.json()
                
                if item['id'] in comments:
                    comments[item['id']].append(item_json)
                else:
                    comments[item['id']] = []
                    comments[item['id']].append(item_json)
                
                if 'kids' in item_json:
                    comments_stack.extend(item_json['kids'])
                    
            else:
                break
```

**Question 5.** What is day of the week with the most stories created? Enter one of the following as an answer: Monday, Tuesday, Wednesday, Thursday, Friday, Saturday, Sunday.

```python
import operator

day_list = {}

for item in items:
    time = datetime.fromtimestamp(item['time'])
    time = time.replace(tzinfo=timezone.utc)
    day = time.strftime('%A')
    
    if day in day_list:
        day_list[day] += 1
    else:
        day_list[day] = 1

best_day = max(day_list.items(), key=operator.itemgetter(1))[0]
print(best_day)
```
